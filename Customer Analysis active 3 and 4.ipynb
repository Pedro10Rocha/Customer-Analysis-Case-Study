Show a plot of the response by the total claim amount.

```python
#Check the maximum amount for the 'Total claim amount'.

numerical['total_claim_amount'].max()
```

```python
#Create samples for the total claim amount.
def total_range(i):
        if i <= 500:
            return 500 
        elif i <= 1000:
            return 1000
        elif i <= 1500:
            return 1500
        elif i <= 2000:
            return 2000
        elif i <= 2500:
            return 2500
        elif i > 2500:
            return 2893
        
df['new_total_claim_amount'] = list(map(total_range,df['total_claim_amount']))
```

```python
#See the correlation between the response and the total claim amount.
sns.countplot(x='new_total_claim_amount', hue='response', data=df)
plt.xticks(ticks=np.arange(6), labels=['0 to 500', '501 to 1000', '1001 to 1500', '1501 to 2000', '2001 to 2500', '2501 to 2893'])
plt.xticks(rotation=90)
```
